```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

```python
mpl.rcParams['axes.spines.left'] = True
mpl.rcParams['axes.spines.right'] = True
mpl.rcParams['axes.spines.top'] = True
mpl.rcParams['axes.spines.bottom'] = True
```

```python
data = pd.read_csv("EPL_Set.csv") # Reading FIFA Dataset
```

```python
Season = data.iloc[5684 : :].Season.unique()[1:] # Filter out ten seasons 2007-08 to 2017-18 name
```

```python
Seasons =[] #  Create am empty season list
for i in Season:  # Looping through above ten season name list
    Seasons.append(i+".csv") # append all the Season name in Seasons with a.csv text
```

```python
def read(file): # Create a function to read csv files
    return pd.read_csv(file).head(4) # show first 4 teams only
```

```python
for s in Seasons: 
    plt.figure(figsize=(12,6)) # Graph Size
    bar = plt.bar(read(s).Club,read(s).Pts,color="#42858C",linewidth=3,edgecolor='black',hatch="/") # Graph Plotting
    plt.xticks([]) # Remove x labels
    plt.yticks([]) # Remove Y lables
    plt.bar_label(bar, labels= read(s).Pts,label_type='edge' ,weight="bold",size=18) # labelling Points 
    plt.bar_label(bar, labels= read(s).Club,label_type='center',weight='bold',size=25,\
                  rotation=90,color='white') # Labelling Teams Name
    
    plt.ylim([0, 110]) # Set Y limit to 100
    plt.xlabel("TOP 4 QUALIFYING TEAMS",weight='bold',size=25) # Set Bottom Label 
    plt.title("EPL " + s[:7], weight='bold',size=25 ) # Set Top Label
    

    plt.show()
```
